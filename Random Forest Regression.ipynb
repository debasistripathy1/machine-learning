```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv("/Users/dev/Desktop/python1/Internshala/Positions_salaries.csv")
```

```python
x=dataset.iloc[:,1].values
x=x.reshape(-1,1)
print(x)
```

```python
y=dataset.iloc[:,2].values
print(y)
```

```python
from sklearn.ensemble import RandomForestRegressor as rfr
regressor= rfr(n_estimators=10);
regressor.fit(x,y)
estimator = regressor.estimators_[9]
```

```python
y_out=regressor.predict([[6]])
print(y_out)
```

```python
x_grid=np.arange(min(x),max(x),0.1)
print(x_grid)
```

```python
x_grid=x_grid.reshape(-1,1)
print(x_grid)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x_grid,regressor.predict(x_grid),color="green")
plt.show()
```
